```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop=stopwords.words('english')
```

```python
import string
```

```python
punctuations=list(string.punctuation)
stop+=punctuations
```

```python
from nltk import pos_tag
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
from nltk.corpus import wordnet
```

```python
def getpos_tag(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
clean_data=[]
for i in range(2000):
    arr=[]
    for word in data[i][0]:
        if word.lower() not in stop:
           # pos_tag=getpos_tag(wo)
            #print(pos_tag)
            pt=pos_tag([word])
            #print(pt)
            Lemmatizer=WordNetLemmatizer()
            w=Lemmatizer.lemmatize(word,pos=getpos_tag(pt[0][1]))
            arr.append(word)
    clean_data.append((arr,data[i][1]))
#clean_data
```

```python
clean_words=[]
for docs in clean_data:
    clean_words+=docs[0]
clean_words
```

```python
freq=nltk.FreqDist(clean_words)
feature_set=freq.most_common(3000)
feature_set=[feature_set[i][0] for i in range(3000)]
dictionary={}
for i in range(3000):
    dictionary[feature_set[i]]=False
dictionary
```

```python
training_data=[]
for docs in clean_data:
    dictionary={}
    for i in range(3000):
        dictionary[feature_set[i]]=False
    for d in docs[0]:
        if d in dictionary:
            dictionary[d]=True
    training_data.append((dictionary,docs[1]))
```

```python
from nltk import NaiveBayesClassifier
```

```python
from sklearn import model_selection
```

```python
Training,Testing=model_selection.train_test_split(training_data)
```

```python
#testing_data=[X_Test,Y_Test]
```

```python
clf=NaiveBayesClassifier.train(Training)
```

```python
nltk.classify.accuracy(clf,Testing)
```
